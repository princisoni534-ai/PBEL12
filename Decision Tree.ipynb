# Understand structure of our decision tree



# Tree:
Real TREE: it has -> Roots -> trunk -> branches -> leaves
Decision Tree: Roots -> Questions -> More questions -> final answers
A decision tree grows exactly like a natural tree, only difference is that instead of growing leaves it grows decisions.
# parent Node:
Parent node creates one or more child nodes
Salary(parents node) -> yes or no -> child node
# child node:
Every node created from another node is called a child node.

Root node -> Internal node -> branch-> child node-> leaf node

- This entire structure is called a Decision Tree!

# Pure vs Impure nodes
- Basket A: u hv apple, apple, apple, apple
- Pure

- Basket B:apple, banana, apple, orange, banana, apple
- Impure


- Tree loves pure nodes

- tree hate impure nodes

- the goal is simple convert impure -> less impuure-> almost pure -> completely pure

- Decision tree grow by increasing purity.

### Entropy
- Imagine you are playing a quetioning game.
- you hv a box

- Entropy measure "How much confusion or uncertainity exists in the data"
- Less confusion -> Low Entropy
- More confusion -> High entropy
- Decision tree always try to Reduce confusion.

Note :

# Gini Index (The Fast Decision Maker)

• Imagine there are two judges in a competition.

• Judge A:

  • Think very carefully

  • Check every rule

  • Overall calculate everything

  • Finally gives the winner.

• Judge B:

  • Looks quickly

  • Makes a fast decision.

  • Still gives almost the same winner.

• Who finishes first?

  • Judge B

• This is exactly the difference between Entropy and Gini Index.

 
• Both measure Impurity
• but Gini Index computationally faster.
• That is why the CART algorithm uses GINI INDEX.

#### Gini index?

- Gini index measures "How impure or mixed a dataset is"
- More mixed data ---->>> Higher Gini
- More Pure data ---->> Lower Gini
- Decision tree always try to Reduce GINI

• Entropy ask "How much information is missing?"
• Gini ask "How likely am I to classify incorrectly?"
• Both reached almost the same decision.
• But Gini requires fewer calcu.
• Therefore, Gini is faster.

Baskets A :

- Basket A: apple, apple apple apple

- ONLY APPLE

- mistake?

- Almost NEVER
- Gini = 0
- Perfectly Pure.

Basket B: Apple Apple banana banana apple banana
- Half Apples
- Half Banana
- Now mistakes can be possible
- Gini Higher

Basket C: Apple, banana grapes watermelons oranges
- everything mixed
- prediction becomes difficult
- Highest impurity

binary classification:
gini = 1 - (p1² + p2²)

p1 = probability of class 1
p2 = probability of class 2

for multiple classes
• for multiple classes

• gini = 1 - ∑(pi^2) 
      # for  sum symbols

dataset : 10 students

- pass - 10

- fails - 0

-probability of pass = 10 / 10 =  1

-probability of fail = 0 / 10 = 0

- gini = 1 - (1 + 0)

- gini = 0

- perfectly pure datasets


- pass = 5
- fail = 5
- pro p = 0.5
- pro f = 0.5
- gini = 1 - (0.25+0.25)
- gini = 0.5
- Maximum gini for binary classification

- customer 100
- purchased - 80
- Not purchased - 20 prob purchase - 0.8 prob of n purchased  - 0.2

gini = 1-(0.64+0.04)gini = 0.32

dataset is fairly pure


```python
def gini(p, n):
    t = p + n
    p1 = p / t
    p2 = n / t

    g = 1 - (p1**2 + p2**2)
    return g

print("pure gini: ", gini(10, 0))
print("balanced gini: ", gini(5, 5))
print("80-20 gini: ", round(gini(8, 2), 3))
```

## Load Dataset:

```python
 
import pandas as pd
from sklearn.datasets import load_breast_cancer

# Load the dataset
cancer = load_breast_cancer()

# Create DataFrame
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)

# Add the target column
df["target"] = cancer.target
df.head(10)
```

```python
print(df.shape)
```

```python
print(df["target"].value_counts)
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

### Train test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print("Training Shape: ", X_train.shape)
```

```python
print("Testing shape:", X_test.shape)
```

## Decision Tree Model 

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(
    criterion="gini",
    random_state=42
)

model.fit(X_train, y_train)
```

### Predict

```python
# Prediction
y_pred = model.predict(X_test)

# Comparison
comparison = pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
})

print(comparison.head())
```

Accuracy 

```python

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

Precision, Recall , F1 

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Precision
precision = precision_score(y_test, y_pred)
print("Precision:", round(precision,2))

# Recall
recall = recall_score(y_test, y_pred)
print("Recall:", round(recall,2))

```

```python
f1 = f1_score(y_test,y_pred)
f1
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Malignant", "Benign"],
            yticklabels=["Malignant", "Benign"])

plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```
