# Understand structure of our decision tree



# Tree:
Real TREE: it has -> Roots -> trunk -> branches -> leaves
Decision Tree: Roots -> Questions -> More questions -> final answers
A decision tree grows exactly like a natural tree, only difference is that instead of growing leaves it grows decisions.
# parent Node:
Parent node creates one or more child nodes
Salary(parents node) -> yes or no -> child node
# child node:
Every node created from another node is called a child node.

Root node -> Internal node -> branch-> child node-> leaf node

- This entire structure is called a Decision Tree!

# Pure vs Impure nodes
- Basket A: u hv apple, apple, apple, apple
- Pure

- Basket B:apple, banana, apple, orange, banana, apple
- Impure


- Tree loves pure nodes

- tree hate impure nodes

- the goal is simple convert impure -> less impuure-> almost pure -> completely pure

- Decision tree grow by increasing purity.

### Entropy
- Imagine you are playing a quetioning game.
- you hv a box
- inside: dog or cat
- can you tell me whats inside?
- DOG....
- You already know the answer .
- THERE IS NO UNCERTAINITY.

- Imagine another box 
- It can contain 
- Inside : Cat or Dog or Rabbit or Horse or Fish or Bird
- Can you guess now ?
- cuz there are many possibilities
- the uncertainity has increased
- the uncertainty is called ENTROPY

- Entropy measure "How much confusion or uncertainity exists in the data"
- Less confusion -> Low Entropy
- More confusion -> High entropy
- Decision tree always try to Reduce confusion.

Ex - suppose there are 55 students

- 55 passed
- 0 failed
- result?
- Entropy = Very Low

ex- 55 students 
- 27 passed 
- 28 failed 
- prediction become difficult 
- Entropy = High 

Ex - 55 students 
- 53 passed 
- 2 failed
- predict: Passed.
- entropy : Mixed data -> Low

Note :
 - More mixed data -> High entropy 
 - More organized Data -> Lower Entropy

Information Gain (How Decision Tree choose the best Question)

- Imagine you are a Detective 
- A theft has occured in a house .

1. was the door locked 
2. was anyone inside?
3. fingerprints found?
4. CCTV available?
5. ques to ask people?

 
- Should you ask this question randomly??? NO
- Because we want to ask the MOST useful information first.
- Decision tree works exactly the same way.
- Which question removes the maximum confusion, that reduction in confusion is called "Information Gain".

- How much uncertainty is removed after splitting the data.

 
- Ex: suppose: 20 students
- 10 passed
- 10 failed
- entropy => High
- Now divide students based on attendance.

- Group A:
- Att > 80%
- 10 passed
- 2 failed

- Group B:
- att <=80
- 0 passed
- 8 failed

- both group are much cleaner now
 

Now divide students based on Roll number.

Group A
- 5 passed
- 5 Failed

Group B
- 5 passed
- 5 failed

Nothing changed
- Prediction is still difficult
- Information Gain is almost zero

- Good Question ->
- Creates pure groups ->
- Entropy Decrease ->
- information Gain increase

Decision tree always selects the question with maximum information gain.
