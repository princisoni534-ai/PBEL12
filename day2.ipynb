```python
#data type 
#list
my_list = [1,2,3,4,5]
#Methods
print(my_list)
my_list.append(6) #add an element to the end of the list
print(my_list) #output:[1,2,3,4,5,6]

my_list.insert(2,10)#insert 10 at index 2
print(my_list)#output :[1,2,10,3,4,5,6]

my_list.remove(3)  # remove by value
print(my_list)     # print list

my_list.pop()      # remove last element
print(my_list)     # print list

#properties
#ordered and mutable
my_list[2]= 20#modifies the element at index 2 
print(my_list) #output:[1,2,20,4,5]

#heterogeneous &duplicate allowed
zen=[1,2,3,4,5,1,2,3,"ai",True,False]
print(zen) #output : [1,2,3,4,5,1,2,3,"ai",True,False]
 
```

```python
# Class
class Car:
    # Constructor
    def __init__(self, color, model, year):
        self.color = color
        self.model = model
        self.year = year

    # Methods
    def start(self):
        print("Car is starting")

    def stop(self):
        print("Car is stopping")

    def play(self, song):
        print(f"Playing {song}")


# Object
bmw = Car("red", "X5", 2020)

# Access attributes
print(bmw.color)
print(bmw.model)
print(bmw.year)

# Call methods
bmw.start()
bmw.play("any")
bmw.stop()
```

```python
#aryan
#create a clas box  , having some
class Box:
    def __init__(self, size, shape, color, volume):
        self.size = size
        self.shape = shape
        self.color = color
        self.volume = volume


# Stack (list)
stack = []

# Push objects into stack
stack.append(Box("Large", "Cube", "Red", 100))
stack.append(Box("Medium", "Rectangle", "Blue", 80))
stack.append(Box("Small", "Square", "Green", 50))

print("Boxes pushed into stack")

# Pop object from stack
box = stack.pop()

print("Popped Box Details:")
print("Size:", box.size)
print("Shape:", box.shape)
print("Color:", box.color)
print("Volume:", box.volume)
```

```python
#While loop use for while

while True:
    try:
        age = int(input("Enter your age: "))

        if age < 13:
            print("You are a child")
        elif age < 18:
            print("You are a teenager")
        else:
            print("You are an adult")

        break  # sahi input milne par loop band

    except ValueError:
        print("Please enter a valid number!")
```
