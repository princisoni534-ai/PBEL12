```python
#merge and join
import pandas as pd

df = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "Name": ["Anand", "Mango", "Pen", "Mobile"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "weight": [55, 62, 48, 70, 60]
})

result = pd.merge(df, df2, how="right",on="id")
print(result)
```

```python
result = pd.merge(df, df2, how="right",on="id")
print(result)
```
